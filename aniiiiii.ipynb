```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
varu=pd.read_csv('energy.csv')
print(varu)
srh=varu[varu['State']=='Andhra']
rcb=varu[varu['State']=='Karanataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])

srh=varu[varu['State']=='telengana']
rcb=varu[varu['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'])
diag[1].bar(rcb['Date'],rcb['Consumption'])

cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]
srh=varu[varu['State']=='Kerala']['Consumption'],sum()
rcb=varu[varu['State']=='Delhi']['Consumption'],sum()
ysr=[srh,rcb]
fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```

```python
import pandas as naga
import matplotlib.pyplot as ani
import seaborn as cse
ntr=naga.read_csv('heart.csv')
print(ntr)
cse.scatterplot(x='age',y='bp',data=ntr,hue='gender')
ani.title('hear attack')
ani.xlabel('age')
ani.ylabel('bp')
ani.show()
cse.lineplot(x='age',y='cholestrol',data=ntr)
ani.show()
cse.boxplot(x='target',y='value',data=ntr)
ani.show()
color=['red','yellow']
cse.countplot(x='target',data=ntr,palette=color)
ani.show()


```
